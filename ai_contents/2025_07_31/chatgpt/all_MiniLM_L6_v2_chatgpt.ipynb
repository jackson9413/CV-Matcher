```python
from sentence_transformers import SentenceTransformer, util

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Input job description and CVs
job_description = "Looking for a data scientist with experience in machine learning, SQL, and Python."
cvs = [
    "Experienced data scientist with strong Python and SQL skills, and deep learning background.",
    "Junior developer with React and frontend experience.",
    "ML engineer with background in big data, PySpark, and cloud services."
]

# Encode
job_vec = model.encode(job_description, convert_to_tensor=True)
cv_vecs = model.encode(cvs, convert_to_tensor=True)

# Compute similarity
similarities = util.cos_sim(job_vec, cv_vecs)[0]

# Rank and print results
for idx in similarities.argsort(descending=True):
    print(f"Score: {similarities[idx]:.4f} | CV: {cvs[idx]}")
```
